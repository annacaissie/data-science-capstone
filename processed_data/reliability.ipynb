```python
# reliability.py

import pandas as pd

reliability_input = "../data/reliability/MBTA Bus, Commuter Rail, & Rapid Transit Reliability.csv"
reliability = pd.read_csv(reliability_input)
```

```python
print(reliability.shape)
```

```python
print(reliability.columns)
```

```python
reliability['service_date'] = pd.to_datetime(reliability['service_date']).dt.date

print(reliability['service_date'].head(10))
```

```python
print(reliability['gtfs_route_short_name'].unique())
```

```python
import pandas as pd

def simplify_route(line):
    if pd.isna(line):
        return None
    elif 'Silver Line' in line:
        return 'Silver Line'
    elif 'Green Line' in line:
        return 'Green Line'
    else:
        return line


reliability['line'] = reliability['gtfs_route_long_name'].apply(simplify_route)
print(reliability['line'].unique())
```

```python
simpl_reliability = reliability.groupby(['service_date', 'line'], as_index=False).agg({
    'otp_numerator': 'sum',
    'otp_denominator': 'sum'
})
```

```python
print(simpl_reliability.shape)
```

```python
simpl_reliability['otp_score'] = simpl_reliability['otp_numerator']/simpl_reliability['otp_denominator']
```

```python
print(simpl_reliability['otp_score'].head(10))
```

```python
simpl_reliability.to_csv('../extra_files/reliability.csv', index=False)
```
